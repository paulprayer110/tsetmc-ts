```python
import pandas as pd
from time import sleep
from bs4 import BeautifulSoup as bs
```

```python
arad = pd.read_excel('D:\interbank\cbportfo14000228.xls')
counter = arad[0]
href = []
for i in counter:
    j = f"http://www.tsetmc.com/Loader.aspx?ParTree=151311&i={i}"
    href.append(j)
arad['href'] = href
arad.columns = ['counter', 'name', 'value', 'percentage', 'href']
arad.drop('counter', inplace=True, axis=1)
arad.to_csv('D:/interbank/aradhrefs14000228.csv')
```

```python
aradlist = pd.read_csv('/content/gdrive/MyDrive/interbank/aradhrefs14000228.csv')
omdehref = []
for i in aradlist['name']:
    try:
        wd.get('http://www.tsetmc.com/Loader.aspx?ParTree=15')
        wd.find_element_by_xpath("""//*[@id="search"]""").click()
        wd.find_element_by_xpath("""//*[@id="SearchKey"]""").send_keys(i)
        sleep(2)
        x = wd.find_element_by_xpath('//*[@id="SearchResult"]/div/div[2]/table/tbody/tr/td[3]/a[3]').get_attribute('href')
        omdehref.append(x)
        sleep(2)
    except:
        print('%s failed' %i)
driver.quit()
```
